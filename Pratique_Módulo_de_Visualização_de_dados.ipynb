```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


df = pd.read_csv('ecommerce_preparados.csv')
print(df.head().to_string())
print(df.tail().to_string())

#Histogrma da Nota
plt.figure(figsize=(10, 6 ))
plt.hist(df['Nota'], bins=10, color='blue', alpha= 0.8)
plt.title('Histograma da Nota')
plt.xticks(ticks=np.arange(2, 5.5, 0.3))
plt.grid(True)
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()

#Gráfico de dispersão
plt.scatter(df['Preço'], df['Desconto'])
plt.title('Gráfico de Dispersão Preço x Quantidade vendida')
plt.xlabel('Preço')
plt.ylabel('Qtd_Vendidos_Cod')
plt.show()

plt.hexbin(df['Preço'], df['Desconto'], gridsize=4, cmap='Blues')
plt.colorbar(label='Preço')
plt.xlabel('Preço')
plt.ylabel('Desconto')
plt.title('Dispersão de Preço e Desconto')
plt.show()

#Mapa de calor
df_corr = df[['Preço', 'Nota', 'Qtd_Vendidos_Cod','Marca_Freq','Material_Freq']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(df_corr, annot=True, fmt='.2f')
plt.title('Mapara de Calor da Correlação entre Variáveis')
plt.show()

#Gráfico de barra

#Gráfico de pizza

#Gráfico de densidade

#Gráfico de Regressão
```
